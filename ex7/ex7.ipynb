K-means

```python
def find_closest_centroids(X, centroids):
    m = X.shape[0]
    k = centroids.shape[0]
    idx = np.zeros(m)

    for i in range(m):
        mid_dist = 1000000
        for j in range(k):
            dist = np.sum((X[i, :] - centroids[j, :]) ** 2)
            if dist < mid_dist:
                mid_dist = dist
                idx[i] = j

    return idx
```

```python
data = loadmat('ex7data2.mat')
X = data['X']
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])

idx = find_closest_centroids(X, initial_centroids)
idx[0:3]
```

```python
def compute_centroids(X, idx, k):
    m, n = X.shape
    centroids = np.zeros((k, n))

    for i in range(k):
        indices = np.where(idx == i)
        centroids[i, :] = (np.sum(X[indices, :], axis=1) / len(indices[0])).ravel()

    return centroids
```

```python
compute_centroids(X, idx, 3)
```

```python
def run_k_means(X, initial_centroids, max_iters):
    m, n = X.shape
    k = initial_centroids.shape[0]
    idx = np.zeros(m)
    centroids = initial_centroids
    for i in range(max_iters):
        idx = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, k)

    return idx, centroids
```

```python
idx, centroids = run_k_means(X, initial_centroids, 10)
```

```python
cluster1 = X[np.where(idx == 0)[0], :]
cluster2 = X[np.where(idx == 1)[0], :]
cluster3 = X[np.where(idx == 2)[0], :]

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(cluster1[:, 0], cluster1[:, 1], s=30, color='r', label='Cluster1')
ax.scatter(cluster2[:, 0], cluster2[:, 1], s=30, color='g', label='Cluster2')
ax.scatter(cluster3[:, 0], cluster3[:, 1], s=30, color='b', label='Cluster3')
ax.legend()
plt.show()
```

```python
def init_centriods(X, k):
    m, n = X.shape
    centroids = np.zeros((k, n))
    idx = np.random.randint(0, m, k)

    for i in range(k):
        centroids[i, :] = X[idx[i], :]

    return centroids
```

主成分分析(PCA)--->principal component analysis

```python
data = loadmat('ex7data1.mat')
data
```

```python
X = data['X']

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
def pca(X):
    X = (X - X.mean()) / X.std()
    X = np.matrix(X)
    cov = (X.T * X) / X.shape[0]
    U, S, V = np.linalg.svd(cov)

    return U, S, V
```

```python
U, S, V = pca(X)
U, S, V
```

```python
def project_data(X, U, k):
    U_reduced = U[:, :k]
    return np.dot(X, U_reduced)
```

```python
Z = project_data(X, U, 1)
Z
```

```python
def recover_data(Z, U, k):
    U_reduced = U[:, :k]
    return np.dot(Z, U_reduced.T)
```

```python
X_recovered = recover_data(Z, U, 1)
X_recovered
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(list(X_recovered[:, 0]), list(X_recovered[:, 1]))
plt.show()
```
